# 8) Итеративный подбор модели
Постановка задачи: будем искать самую оптимальную модель из следующих алгоритмов:
- Деревья принятия решений 
- Cлучайные леса 
- Градиентный бустинг деревьев.

Задача предсавляет собой бинарную классификацию (предсказание 0 или 1), причем: 0 - malignant(злокачественный), 1- benign(доброкачественный).
Проверку алгоритмов будем реализовывать перекрестной проверкой c перемешиванием.

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfold = KFold(n_splits=5, shuffle=True)
```

```python
# Разбиваем данные на: 1) Обучение 2) Тест
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 0:-1], df.iloc[:, -1], stratify=df.iloc[:, -1],
                                                    random_state=42)
```

# 8.1) Алгоритм: Деревья принятия решений

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
```

```python
# Скармливание данных: 
tree.fit(X_train, y_train) 
# Теперь, в tree хранится модель МО
```

```python
print(f'Правильность на обучающем наборе:\n {tree.score(X_train, y_train)}')
print(f'Правильность на тестовом наборе:\n {tree.score(X_test, y_test)}')
```

```python
# Перекрестная проверка:
result1 = cross_val_score(tree, cancer.data, cancer.target, cv=kfold)
print(f'Значения правильности перекрестной проверки для моделей:\n {result1}')
print(f'Среднее значение правильности перекрестной проверки:\n {np.average(result1)}')
```

# 8.2) Алгоритм: Случайный лес

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=5, random_state=2)
```

```python
# Скармливание данных: 
random_forest.fit(X_train, y_train)
# Теперь, в random_forest хранится модель МО
```

```python
print(f'Правильность на обучающем наборе:\n {random_forest.score(X_train, y_train)}')
print(f'Правильность на тестовом наборе:\n {random_forest.score(X_test, y_test)}')
```

```python
# Перекрестная проверка:
result2 = cross_val_score(random_forest, cancer.data, cancer.target, cv=kfold)
print(f'Значения правильности перекрестной проверки для моделей:\n {result2}')
print(f'Среднее значение правильности перекрестной проверки:\n {np.average(result2)}')
```

# 8.3) Алгоритм: Градиентный бустинг деревьев

```python
from sklearn.ensemble import GradientBoostingClassifier
gradient_boosting = GradientBoostingClassifier(random_state=0)
```

```python
# Скармливание данных:
gradient_boosting.fit(X_train, y_train)
# Теперь, в gradient_boosting хранится модель МО 
```

```python
print(f'Правильность на обучающем наборе:\n {gradient_boosting.score(X_train, y_train)}')
print(f'Правильность на тестовом наборе:\n {gradient_boosting.score(X_test, y_test)}')
```

```python
# Перекрестная проверка:
result3 = cross_val_score(gradient_boosting, cancer.data, cancer.target, cv=kfold)
print(f'Значения правильности перекрестной проверки для моделей:\n {result3}')
print(f'Среднее значение правильности перекрестной проверки:\n {np.average(result3)}')
```

# 9) Выбор оптимальной модели по результатам тестов для данного датасета

На основании вышеприведенных тестов (1.train_test_split  2.перекрестная проверка), выбираем модель МО gradient_boosting, алгоритмом модели является Градиентный бустинг деревьев.

```python
optimal_model = gradient_boosting
```

# 10) ROC - AUC для оптимальной модели

```python
pred = optimal_model.predict(X_test)
print(pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion = confusion_matrix(y_test, pred)
print(f'Матрица ошибок:\n{confusion}')
```

```python
import mglearn
mglearn.plots.plot_binary_confusion_matrix()
```

```python
from sklearn.metrics import precision_score
print(f'Точность оптимальной модели: {precision_score(y_test, pred)}')
```

```python
from sklearn.metrics import recall_score
print(f'Полнота оптимальной модели: {recall_score(y_test, pred)}')
```

```python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, optimal_model.decision_function(X_test))
```

```python
close_zero = np.argmin(np.abs(thresholds))
plt.plot(precision[close_zero], recall[close_zero], 'o', markersize=10,
label="порог 0", fillstyle="none", c='k', mew=2)
plt.plot(precision, recall, label="кривая точности-полноты")
plt.title('ROC-кривая оптимальной модели')
plt.xlabel("Точность (FPR)")
plt.ylabel("Полнота (TPR)")
plt.legend(loc="best")
plt.show()
```

Известно, что чем ближе подходит кривая к верхнему правому углу, тем лучше классификатор. Точка в верхнем правом углу означает высокое значение точности и высокое значение полноты для соответствующего порога.

```python
from sklearn.metrics import roc_auc_score
optimal_model_auc = roc_auc_score(y_test, optimal_model.decision_function(X_test))
print(f'AUС для оптимальной модели: {optimal_model_auc}')
```
